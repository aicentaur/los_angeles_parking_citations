```python
# import dependencies
import os
import csv
from pandas import DataFrame
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
# read csv into data frame
meters_inventory = pd.read_csv('meters_inventory.csv', sep=",")
meters_inventory
```

```python
# check data type in dataframe convert data to an appropriate format
meters_inventory.dtypes
```

```python
# Create database connection
connection_string = "root:root@127.0.0.1:3306/la_parking_db"
engine = create_engine(f'mysql+pymysql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
# Load DataFrames into database
meters_inventory.to_sql(name = 'meters_inventory', con = engine, if_exists = 'append', index = False) 
```
