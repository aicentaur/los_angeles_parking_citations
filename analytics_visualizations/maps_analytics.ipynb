```python
# import dependencies
%matplotlib notebook
import matplotlib.pyplot as plt
import os
import csv
from pandas import DataFrame
import pandas as pd
import numpy as np
```

# Graph 1 | Citations per Year/Month

```python
# load data
path = 'visualization_data/graph_1_citations_by_yyyy.csv'
graph_1 = pd.read_csv(path, sep=",")

# plot data
graph_1.plot.bar(x='Year', y='counts', label = "Parking Citations")

#format plot
plt.title("Parking Citations by Year", fontdict = {"fontsize": 14, "fontweight": "bold"})
plt.xlabel("Years", fontdict = {"fontsize": 9, "fontweight": "bold"})
plt.ylabel("Number of Parking Citations (in Millions)", fontdict = {"fontsize": 9, "fontweight": "bold"})

# Save the figure
plt.tight_layout()
plt.savefig("graphs/citations_by_year.png")

# Show the figure
plt.show()
```

# Map Visualization 1 | Areas with Highest Number of Tickets

```python
# load data and create "tuple_coordinates" column
map_1 = pd.read_csv('visualization_data/map_1_citations_areas.csv')
map_1['tuple_coordinates'] = list(zip(map_1.longitude, map_1.latitude))
map_1.head(3)
```

```python
# put "tuple_coordinates" column into list
list_tuple_1 = map_1['tuple_coordinates'].tolist()
print(f"Total number of records:{len(list_tuple_1)}")
```

```python
from ipyleaflet import Map, Marker, MarkerCluster, FullScreenControl

m = Map(center=(34.0522, -118.2437), zoom=10)


markerList = []
for i in list_tuple_1:
    marker = Marker(location=(i[1], i[0]))
    markerList.append(marker)
    

marker_cluster = MarkerCluster(
markers=(markerList))

m.add_layer(marker_cluster);
m.add_control(FullScreenControl())

m
```

# Graph 5 | Fine Amount Distribution

    `$505` (5 tickets total, with no violation type)<br>
    `$363` (disabled and handicap related violations)
