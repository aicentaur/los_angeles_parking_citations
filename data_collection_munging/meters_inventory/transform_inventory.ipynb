```python
# import dependencies
import json
import os
import csv
from pandas import DataFrame
import pandas as pd
import numpy as np
```

```python
data_list = []
for x in range(1, 35):
    filename = "api_responses/"f"{x:010}.json"
    with open(filename) as json_file:  
        data = json.load(json_file)
        for s in data:
            data_list.append(s)
    

transformed_df = pd.DataFrame(data_list)
transformed_df
```

```python
transformed_df.info()
```

```python
transformed_df.rename(columns =
                      {"blockface": "address",
                       "metertype": "meter_type",
                       "parkingpolicy": "parking_policy",
                       "raterange": "rate_range",
                       "ratetype": "rate_type",
                       "spaceid": "meter_id",
                       "stcl": "street_cleaning",
                       "timelimit": "parking_limit",},
                      inplace=True)
```

```python
transformed_df.drop(
    [":@computed_region_2dna_qi2s",
     ":@computed_region_k96s_3jcv",
     ":@computed_region_kqwf_mjcx",
     ":@computed_region_qz3q_ghft",
     ":@computed_region_tatf_ua23",
     ":@computed_region_ur2y_g4cx",],
    axis = 1, inplace = True)
```

```python
# convert the dictionary in latlng column with latitude and longitude values to two different columns
transformed_df['latlng'].apply(pd.Series)
```

```python
# combine converted data with the rest of the dataframe
transformed_df_v2 = pd.concat([transformed_df.drop(['latlng'], axis=1), transformed_df['latlng'].apply(pd.Series)], axis=1)
```

```python
# calculate the length of each parking policy record and add the result to a separate column
transformed_df_v2["policy_complexity"] = transformed_df_v2["parking_policy"].apply(len)
```

```python
# extract numerical value from "parking_limit" column and put result in a separate column
transformed_df_v2["park_limit"] = transformed_df_v2["parking_limit"].str.extract('(\d+)').astype(int)
```

```python
# convert minutes into hours in "park_limit" column and assign results to the same column
transformed_df_v2["park_limit"] = np.where(transformed_df_v2["park_limit"]>=10, transformed_df_v2["park_limit"]/60, transformed_df_v2["park_limit"])
                                                                                     
```

```python
transformed_df_v2[["latitude", "longitude"]] = transformed_df_v2[["latitude", "longitude"]].apply(pd.to_numeric, errors = "coerce")
```

```python
# Create a filtered dataframe from specific columns
name_col = ["address", "latitude", "longitude", "meter_type", "parking_policy", "policy_complexity", "rate_range", "rate_type", "meter_id", "street_cleaning", "park_limit", ]
db_ready_df = transformed_df_v2[name_col].copy()
db_ready_df
```
